## Pergunta de Negócio 1:

### Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

```python
df_dsa[df_dsa['Categoria'] == 'Office Supplies'].groupby('Cidade')['Valor_Venda'].sum().sort_values(ascending=False).index[0]
```

## Pergunta de Negócio 2:

### Qual o Total de Vendas Por Data do Pedido?

Demonstre o resultado através de um gráfico de barras.

```python
df_dsa['Data_Pedido'] = pd.to_datetime(df_dsa["Data_Pedido"], format="%d/%m/%Y")
```

```python
df_dsa_2 = df_dsa.groupby('Data_Pedido')['Valor_Venda'].sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (20,5))
df_dsa_2.plot(x='Data_Pedido', y='Valor_Venda', title='Total de Vendas por Data do Pedido')
plt.show()
```

## Pergunta de Negócio 3:

### Qual o Total de Vendas por Estado?

Demonstre o resultado através de um gráfico de barras.

```python
plt.figure(figsize=(20,5))
df_dsa.groupby('Estado')['Valor_Venda'].sum().plot(kind='bar', title='Total de Vendas por Estado');
```

## Pergunta de Negócio 8 (Desafio Nível Master):

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
df_dsa_copy = df_dsa.copy()
media_anterior = df_dsa_copy['Valor_Venda'].mean()
df_dsa_copy['Valor_Venda'] = np.where(df_dsa_copy['Valor_Venda']>1000, df_dsa_copy['Valor_Venda']*0.85, df_dsa_copy['Valor_Venda'])
media_posterior = df_dsa_copy['Valor_Venda'].mean()

print(f"Média Anterior: {media_anterior}\nMedia Posterior: {media_posterior}")
```

## Pergunta de Negócio 9 (Desafio Nível Master Ninja):

### Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?

Demonstre o resultado através de gráfico de linha.

```python
df_dsa['Mes'] = df_dsa['Data_Pedido'].dt.month
```

```python
df_dsa.head()
```

```python
import seaborn as sns

df_dsa4 = df_dsa.groupby(['Segmento', 'Ano', 'Mes'])['Valor_Venda'].mean()
df_dsa4.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Média de Vendas por Segmento, Ano e Mês')
plt.xlabel('Segmento, Ano e Mês')
plt.ylabel('Média de Vendas')
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
# sns.lineplot(data=df_dsa4, x=['Ano', 'Mes'], y='Valor_Venda', hue='Segmento')
```
