```python
import pandas as pd

df = pd.read_table('https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro04/oficina/arquivos/dadosoficina04.txt', sep=',')
```

```python
df.head()
```

<font color='blue'>**2. Crie uma visualização que mostre a proporção de vendas global entre gêneros de jogos da empresa Nintendo;**</font>

```python
total_vendas_por_genero_nintendo = df[df['Publisher'] == 'Nintendo'].groupby('Genre')['Global_Sales'].sum()
```

```python
total_vendas_nintendo = df.loc[df['Publisher'] == 'Nintendo','Global_Sales'].sum()
```

```python
df2 = total_vendas_por_genero_nintendo/total_vendas_nintendo
```

```python
df2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(7,7))
df2.plot.pie(autopct='%.2f%%')
plt.ylabel('Vendas Global')
plt.title('Proporção de Vendas Global entre Gêneros de Jogos da Empresa Nintendo')
plt.show()
```

<font color='blue'>**3. Crie uma visualização que mostre a proporção de quantidade de jogos por plataforma a partir do ano de 2015.**</font>

```python
total_jogos_maior_2015 = df.loc[df['Year']>=2015,'Platform'].count()
```

```python
total_jogos_por_plataforma_maior_2015 = df[df['Year']>=2015].groupby('Platform')['Platform'].count()
```

```python
total_jogos_por_plataforma_maior_2015
```

```python
proporcao = total_jogos_por_plataforma_maior_2015/total_jogos_maior_2015
```

```python
proporcao
```

```python
proporcao.plot.pie(autopct='%.2f%%')
plt.title('Proporção de quantidade de jogos por plataforma a partir do ano de 2015')
plt.ylabel('Plataforma')
plt.show()
```
