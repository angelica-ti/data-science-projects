```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

df = pd.read_json('dadosoficina03.txt')
df.head()
```

**<font color='blue'>2. Crie uma visualização que mostre o orçamento de todos os filmes do conjunto de dados. Qual o filme com maior orçamento e com menor orçamento?</font>**

```python
plt.figure(figsize=(10,10))
sns.barplot(df, x='Budget_M', y='Film', orient='h')
plt.title('Orçamento por Filme')
```

```python
df.loc[df['Budget_M'] == df['Budget_M'].min(), 'Film'].iloc[0], df.loc[df['Budget_M'] == df['Budget_M'].max(), 'Film'].iloc[0]
```

```python
df.iloc[df['Budget_M'].idxmin(),:]['Film'], df.iloc[df['Budget_M'].idxmax(),:]['Film'] 
```

```python
df.loc[df['Budget_M'].idxmin(),'Film'], df.loc[df['Budget_M'].idxmax(),'Film'] 
```

**<font color='blue'>3. Crie uma visualização que mostre a bilheteria de todos os filmes do conjunto de dados. Qual o filme com maior bilheteria e com menor bilheteria?</font>**

```python
plt.figure(figsize=(10,10))
sns.barplot(df, x='Worldwide_Gross_M', y='Film', orient='h')
plt.title('Bilheterias por Filme')
```

```python
df.loc[df['Worldwide_Gross_M'].idxmin(), 'Film'], df.loc[df['Worldwide_Gross_M'].idxmax(), 'Film']
```

**<font color='blue'>4. Crie uma visualização que mostre a quantidade de filmes por gênero em cada ano. Em que anos foram feitos menos filmes de ação? E em que ano foram feitos menos filmes de animação?</font>**

```python
df2 = df.groupby(['Year','Genre'])['Film'].count()
```

```python
df2
```

```python
plt.figure(figsize=(10,5))
sns.countplot(df, x='Year', hue='Genre')
plt.title('Quantidade de Filmes por Gênero em cada Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Filmes')
plt.show()
```

**R:** 2008 e 2010 foram feitos menos filmes de ação! 2008 e 2011 foram feitos menos filmes de animação!

<font color='blue'>**5. Crie uma visualização que mostre a média de orçamento de filmes por gênero em cada ano. Em que ano os filmes de ação tiveram maior média de orçamento? E em que ano os filmes de animação tiveram a maior média de orçamento**</font>

```python
df.groupby(['Year','Genre'])['Budget_M'].mean()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(df, x='Year', y='Budget_M', hue='Genre')
plt.title('Média de Orçamento de Filmes por Gênero em Cada Ano')
plt.xlabel('Ano')
plt.ylabel('Média de Orçamento')
plt.show()
```

**R:** Em 2008 os filmes de animação tiveram a menor média de orçamento
